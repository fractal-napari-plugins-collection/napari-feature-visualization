# Interactive feature visualization testing

Uses the same label image and feature dataframe as the pytest setup for interactive exploration.

```python
import napari
from napari_feature_visualization._tests.test_feature_vis import create_feature_df, create_label_img
from napari_feature_visualization.feature_vis import feature_vis
```

```python
lbl_img = create_label_img()
df = create_feature_df()
df
```

```python
viewer = napari.Viewer()
label_layer = viewer.add_labels(lbl_img)
label_layer.features = df

widget = feature_vis()
viewer.window.add_dock_widget(widget)
```
